```python
# Dataset clientes con Cta Sueldo
data_sdo=data2[data2.CASUELDO_TIENE==1]
data_sdo.shape
```

```python
# Dataset clientes Jubilados
data_jub=data2[data2.CAJUBILADOPROV_TIENE==1]
data_jub.shape
```

```python
# TODO: Elegir un soporte mínimo del 10% para que tenga sentido las asociaciones
freq_is = apriori(data_jub, min_support=0.05, use_colnames=True)
# Ordeno por asociaciones mas importantes
freq_is.sort_values(by='support',ascending=False)
```

```python
# Mostrar los conjuntos mayores a 2 items frecuentes
freq_is['length'] = freq_is['itemsets'].apply(lambda x: len(x))
freq_is_2=freq_is[freq_is['length'] > 1]

# Ordeno por asociaciones mas importantes
freq_is_2.sort_values(by='support',ascending=False)
```

```python
# Crear reglas de asociación a partir de los conjuntos de items frecuentes
as_rules = association_rules(freq_is_2, metric="confidence", min_threshold=0.05,support_only=True) 
as_rules.sort_values(by='support',ascending=False)
```

```python
## Eliminación de variables con ruido que no suman valor y son linealmente dependientes

as_rules[
    as_rules["antecedents"].apply(lambda x: (len(x) > 1) & (not sum(["CAJUBILADOPROV" in item for item in x]))) &
    as_rules["consequents"].apply(lambda x: (not sum(["CAJUBILADOPROV" in item for item in x]) ))
].sort_values(by="support", ascending=False).iloc[:20]
```

```python
# Función de predicción de Producto Consecuente en funcion del producto Antecedente

def predict(antecedent, rules, by_sort="support", max_results=6):
    """ Dado un antecedente, conseguir consecuentes ordenados por by_sort"""
    
    # get the rules for this antecedent
    preds = rules[rules['antecedents'] == antecedent].sort_values(by=by_sort, ascending=False)
    
    # a way to convert a frozen set with one element to string
    preds = preds['consequents']
    
    return preds[:max_results]

def antecedent(consequent, rules, by_sort="support", max_results=6):
    """ Dado un consecuente (producto a recomendar), conseguir conjuntos de antecedentes ordenados por by_sort"""
    
    # get the rules for this antecedent
    preds = rules[rules['consequents'] == consequent].sort_values(by=by_sort, ascending=False)
    
    # a way to convert a frozen set with one element to string
    preds = preds['antecedents']
    
    return preds[:max_results]

def users_to_recommend(consequent, set, data):
    """
    Dado un consecuente, un conjunto de antecedentes, buscar a las personas
    que no tengan el consecuente y tengan el antecedente para recomendarles el consecuente.
    """
    to_rec = data[data[consequent].eq(0)]
    for item in iter(set):
        to_rec = to_rec[to_rec[item].eq(1)]
    return to_rec.iloc[:,0]        
```

```python
# EJEMPLO DE CLIENTES A OFRECER TARJETAS DE CRÉDITO EN FUNCIÓN DE SUS PRODUCTOS ACTUALES
print(antecedent({"TCCONVENCIONAL_TIENE"}, as_rules))
# Usuarios que podrían recibir Tarjeta Mastercard (no la tienen) y poseen su primer antecedente por Support (CAD, PP)
users_to_recommend("TCCONVENCIONAL_TIENE", antecedent({"TCCONVENCIONAL_TIENE"}, as_rules).iloc[0], data)
```
